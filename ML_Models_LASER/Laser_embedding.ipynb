```python

# Step 4: Import libraries
import pandas as pd
import numpy as np
from laserembeddings import Laser
import os

# Step 5: Define file paths
input_file = '/content/telugu_reviews_merged.xlsx'
output_folder = '/content/drive/My Drive/ML_End_Sem/Telugu/ML_Models_Laser'
output_file = os.path.join(output_folder, 'telugu_reviews_with_laser.xlsx')

# Step 6: Create output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Step 7: Load the dataset
try:
    df = pd.read_excel(input_file, engine='openpyxl')
    # Keep only 'comment' and 'label' columns
    df = df[['comment', 'label']]
    print(f"Loaded {len(df)} comments from the dataset.")
except Exception as e:
    print(f"Error loading file: {e}")
    raise

# Step 8: Clean the data
# Remove rows with missing or invalid comments/labels
df = df.dropna(subset=['comment', 'label'])
df = df[df['comment'].str.strip() != '']
df = df[df['label'].isin([0, 1])]
print(f"After cleaning, {len(df)} comments remain.")

# Step 9: Initialize LASER
try:
    laser = Laser()
    print("LASER initialized successfully.")
except Exception as e:
    print(f"Error initializing LASER: {e}")
    raise

# Step 10: Generate LASER embeddings
def get_laser_embedding(text):
    try:
        # LASER expects a list of strings
        embedding = laser.embed_sentences([text], lang='te')[0]  # 'te' for Telugu
        return embedding.tolist()  # Convert to list for Excel compatibility
    except Exception as e:
        print(f"Error embedding text '{text[:50]}...': {e}")
        return [0.0] * 1024  # Return zero vector if embedding fails

# Apply embeddings to the comment column
df['embedding'] = df['comment'].apply(get_laser_embedding)
print("Generated LASER embeddings for all comments.")

# Step 11: Verify the embeddings
df['embedding_length'] = df['embedding'].apply(len)
if not all(df['embedding_length'] == 1024):
    print("Warning: Some embeddings do not have the expected length of 1024.")
else:
    print("All embeddings have the correct length of 1024.")

# Step 12: Drop the temporary column
df = df.drop(columns=['embedding_length'])

# Step 13: Save the results to Excel
try:
    df.to_excel(output_file, index=False, engine='openpyxl')
    print(f"Successfully saved output to {output_file}")
except Exception as e:
    print(f"Error saving file: {e}")
    raise

# Step 14: Verify the saved file
if os.path.exists(output_file):
    saved_df = pd.read_excel(output_file, engine='openpyxl')
    print(f"Verified: Output file contains {len(saved_df)} rows.")
else:
    print("Error: Output file was not created.")
```
